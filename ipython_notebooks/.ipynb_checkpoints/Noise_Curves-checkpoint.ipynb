Let's redo this noise analysis:

    Assume we detect the output at port 2, the feedback loop puts the amplified signal back into port 1, and we send in a signal through port 3.
    
Then the output waves ($b$) are given by the S-matrix multiplied by the input waves ($a$):

$$
\vec{b} = S \vec{a}
$$

and if we write that out explicitly,

$$
b_2(t) = S_{21}a_1(t) + S_{23}a_3(t)
$$

Once again, $a_3$ is the injected signal at port 3, and $a_1$ is the amplitude of the feedback going into port 1, so we can write $a_1$ as

$$
a_1(t) = G b_2(t-t_0)
$$

where $G$ is the voltage gain in the loop and $t_0$ is the roundtrip time to go from port 2 to port 1.

We'll make two crucial assumptions:

    Assume the output is steady state:
$$\langle b_2^2(t) \rangle = \langle b_2^2(t-t_0) \rangle = \bar{b}_2^2$$

    Assume the input signal has a coherence time (so the auto-correlation decays exponentially):
$$\langle a_3(t)a_3(t-nt_0)\rangle = \bar{a}_3^2 e^{-nt_0/t_{sig}}$$

If you disagree with any of the conclusions of this analysis, these assumptions are probably what you'll want to return to as
they are the starting points to madness.

Note: Let's choose our units to be convenient: $\bar{b}_2^2$ is the output power, $\bar{a}_3^2$ is the input signal power. For example, if we sent in $KT_{sig}B$ Watts to port 3, $\bar{a}_3^2 = kT_{sig}B$.

Now let's do some math.

Putting in the feedback into the equation for $b_2(t)$, we have

$$b_2(t) = S_{21}Gb_2(t-t_0) + S_{23}a_3(t)$$

Let's try and simplify the notation a little bit so we don't have to type as much: assign new variables to $S_{21}G$ and $S_{23}a_3(t)$:

$$
x = S_{21}G
$$
$$
s(t) = S_{23}a_3(t)
$$

So 

$$b_2(t) = x b_2(t-t_0) + s(t).$$

Squaring and taking the time average, we get the avg. detected output power, $\bar{b}_2^2$:

$$\bar{b}_2^2(t) = x^2\bar{b}_2^2 + \bar{s}^2 + 2x\langle s(t) b_2(t-t_0)\rangle$$

We are competent enough to take the first term over to the LHS:

$$(1 - x^2)\bar{b}_2^2 = \bar{s}^2 + 2x\langle s(t) b_2(t-t_0)\rangle$$

I'm going to expand the second term on the RHS (this is probably where I'll mess everything up...):

$$2x\langle s(t) b_2(t-t_0)\rangle = 2x[\bar{s}^2e^{-t_0/t_{sig}}] + 2x^2 \langle s(t) b_2(t-2t_0)\rangle$$

Going another level of recursion deeper
$$2x\langle s(t) b_2(t-t_0)\rangle = 2x[\bar{s}^2e^{-t_0/t_{sig}}] + 2x^2[\bar{s}^2e^{-2t_0/t_{sig}}] + 2x^3 \langle s(t) b_2(t-3t_0)\rangle$$

So if I'm not crazy, we can express this as

$$2x\langle s(t) b_2(t-t_0)\rangle = \bar{s}^2 \frac{2x e^{-t_0/t_{sig}}}{1-xe^{-t_0/t_{sig}}}$$

and putting that back into the whole equation, we get

$$(1 - x^2)\bar{b}_2^2 = \bar{s}^2\bigg(1 + \frac{2x e^{-t_0/t_{sig}}}{1-xe^{-t_0/t_{sig}}} \bigg)$$

and making it pretty

$$\bar{b}_2^2 = \frac{\bar{s}^2}{1 - x^2}\bigg(\frac{1 + x e^{-t_0/t_{sig}}}{1-xe^{-t_0/t_{sig}}} \bigg)$$

What does this all mean?

    Scenario 1: Noise
    
If we are analyzing noise, it has a coherence time of $t_{sig} = t_{cav}$, and $\bar{s}^2 = kT_{sys}B$ (for port 2 critically coupled), so 

$$P_{noise} = \frac{kT_{sys}B}{1 - x^2}\bigg(\frac{1 + x e^{-t_0/t_{cav}}}{1-xe^{-t_0/t_{cav}}} \bigg)$$

    Scenario 2: Axion Signal
    
If we have an axion signal, it has a coherence time of $t_{sig} = t_a$, and $\bar{s}^2 = P_a$. We don't include the factor of $S_{23}$ because $P_a$ is already defined as the power exiting the cavity:
$$
P_a = \eta g^2B^2VC\text{min}(Q_L, Q_a)\frac{\rho_a}{m_a}
$$
where $\eta$ is the fraction of power that exits the cavity (1/2 for critical coupling). In the regenerative circuit $P_a \propto \text{min}(Q_{enhanced}, Q_a)$; where $Q_{enhanced}$ goes as 

$$Q_{enhanced} = \frac{Q_L}{1-x}$$

(we won't prove it but this is a good reference for seeing why: http://qrp.kearman.com/images/q-mult/Harris_Q-Mult.pdf)
So... the output detected signal power in this scenario goes as 

$$P_{signal} = \frac{P_{a,original}}{1 - x^2}\text{min}\bigg(\frac{Q_{enhanced}}{Q_L}, \frac{Q_a}{Q_L}\bigg)\bigg(\frac{1 + x e^{-t_0/t_{a}}}{1-xe^{-t_0/t_{a}}} \bigg)$$
where $P_{a, original}$ refers to the original axion power when $x=0$; and rewritten, the expression is

$$P_{signal} = \frac{P_{a,original}}{1 - x^2}\text{min}\bigg(\frac{1}{1-x}, \frac{t_{cav}}{t_a}\bigg)\bigg(\frac{1 + x e^{-t_0/t_{a}}}{1-xe^{-t_0/t_{a}}} \bigg)$$

    Scenario 3: CW Signal
    
For an injected test tone that does NOT have a dependence on $Q_{enhanced}$, we get

$$P_{signal} = \frac{|S_{23}|^2 P_{in}}{1 - x^2}\bigg(\frac{1 + x e^{-t_0/t_{sig}}}{1-xe^{-t_0/t_{sig}}} \bigg)$$

In most cases these test tones have a coherence time that is very large, at least $t_{sig} > 1\text{ sec}$ at 2 GHz. Therefore we are in the limit when $t_0 \ll t_{sig}$, so 

$$P_{signal} = \frac{|S_{23}|^2 P_{in}}{1 - x^2}\bigg(\frac{1 + x }{1-x} \bigg) = \frac{|S_{23}|^2 P_{in}}{(1 - x)^2}$$

Finally, let's bring this all together:

    SNR Enhancement
    
For an axion signal, the SNR enhancement due to the feedback loop (i.e., $x$) is

    Axion SNR enhancement
$$
\frac{SNR(x)}{SNR(x=0)} = \text{min}\bigg(\frac{1}{1-x}, \frac{t_{cav}}{t_a}\bigg)\bigg(\frac{1 + x e^{-t_0/t_{a}}}{1-xe^{-t_0/t_{a}}}\bigg)\bigg(\frac{1 - x e^{-t_0/t_{cav}}}{1+xe^{-t_0/t_{cav}}} \bigg)
$$

    CW SNR enhancement
$$
\frac{SNR(x)}{SNR(x=0)} = \bigg(\frac{1 + x e^{-t_0/t_{sig}}}{1-xe^{-t_0/t_{sig}}}\bigg)\bigg(\frac{1 - x e^{-t_0/t_{cav}}}{1+xe^{-t_0/t_{cav}}} \bigg)
$$

There are two big surprising (to me) conclusions one can take away from this:

    Conclusions:
    
    1. You get more enhancement in your axion signal from the feedback loop than just the factor of 1/1-x in the Q increase.
    
    2. The follow up to this is that you get MORE SNR if you start with a low Q and put it in a feedbck circuit, than if you            have a passive resonator with a loaded Q equal to the axion Q from the start. (this is in effect what Matt Dobbs was saying.
    
    3. Just to repeat the point a different way, if you put a resonator with loaded Q equal to the axion Q in a regenerative circuit, there will be ZERO SNR enhancement.
       
Some related conclusions, as we'll see from the plots below, are that

    4. There is no difference in the behavior for delay times after they are larger than the cavity coherence time.
    
    5. if there is no delay, there is no enhancement for a cw signal. (but there is for an axion)

```python
#let's make a plot of the axion signal SNR enhancement w/ and w/o a delay line:
Q_a = 1.e6
Q_cav = 1.e3
omega_0 = 2.*np.pi*2.256*1.e9
t_a = Q_a/omega_0
t_cav = Q_cav/omega_0
t_1 = [2.4e-6, 1.e-9]

num_steps = 50
x_vals = list(np.linspace(0.01, .999, num=num_steps))
y_vals0 = [np.minimum(1./(1.-x), t_a/t_cav)*(1.+np.exp(-t_1[0]/t_a)*x)*(1.-np.exp(-t_1[0]/t_cav)*x)/((1.-np.exp(-t_1[0]/t_a)*x)*(1.+np.exp(-t_1[0]/t_a)*x)) for x in x_vals]
y_vals1 = [np.minimum(1./(1.-x), t_a/t_cav)*(1.+np.exp(-t_1[1]/t_a)*x)*(1.-np.exp(-t_1[1]/t_cav)*x)/((1.-np.exp(-t_1[1]/t_a)*x)*(1.+np.exp(-t_1[1]/t_a)*x)) for x in x_vals]

g = graph.graphxy(width=8,
                  y=graph.axis.log(min=1e-1, max=1e4, title=r"axion SNR enhancement"),
                  x=graph.axis.lin(max=1, title=r"closed loop gain, $t_{cav} \ll t_a$"),
                  key=graph.key.key(pos="tl", dist=0.1))
g.plot([graph.data.values(x=list(x_vals), y=list(y_vals0), title=r"delay"),
        graph.data.values(x=list(x_vals), y=list(y_vals1), title=r"no delay")],
       [graph.style.line([color.gradient.Rainbow])])

g.writePDFfile("delay_no_delay")
#ok, so there is a visible improvemet with the delay! phew.
```
